```r
encuestas<-c("ago-21", "jun-2017","jun-2013")
```

```r
url_base1<-"https://copred.cdmx.gob.mx/storage/app/media/base-de-datoscopred-"
url_base2<-"https://copred.cdmx.gob.mx/storage/app/media/base-de-datos-copred-"
```

```r
#Descargar los datos
for (i in encuestas) {
    if (i == "ago-21") {
        url <- paste0(url_base1, i, ".xlsx")
    } else {
        url <- paste0(url_base2, i, ".xlsx")
    }
    download.file(url, destfile = paste0("microdatos/", i, ".xlsx"),
                    mode = "wb")}
```

```r
##Leer ago-21
edis<-openxlsx::read.xlsx("microdatos/ago-21.xlsx")%>%
janitor::clean_names()
head(edis)
```
